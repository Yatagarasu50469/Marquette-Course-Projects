```python
import sys
from numpy import *
from matplotlib import gridspec
import os
import numpy as np
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# def func(pct, allvals):
#     absolute = int(pct/100.*np.sum(allvals))
#     if pct < 2:
#         fVal = " "
#     else:
#         fVal = "{:.1f}%  ({:d})".format(pct, absolute)
#     return fVal

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    if pct < 2:
        fVal = " "
    else:
        fVal = "{:.1f}%  ({:d})".format(pct, absolute)
    return fVal
```

```python
GAN_v1      = pd.read_excel(r"L_batchSize8_training_epoch5000_truncatedNormal.xlsx") 
CVAE_v1     = pd.read_excel(r"L_batchSize32_training_epoch10000_truncatedNormal.xlsx") 
CVAE_v3     = pd.read_excel(r"L_S_from_Scratch.xlsx")
Val_Set     = pd.read_excel(r"L_S_From_Existing.xlsx") 
```

```python
# L_batchSize8_training_epoch5000_truncatedNormal
# L_batchSize32_training_epoch10000_truncatedNormal

%matplotlib qt5
nrow = 1
ncol = 1

fig = plt.figure(figsize=(5, 5)) 
# fig.suptitle('GAN vs CVAE Analysis', fontsize=24) 


gs = gridspec.GridSpec(nrow, ncol, width_ratios=[1],
         wspace=0.2, hspace=0.3, top=0.92, bottom=0.05, left=0.08, right=0.98) 
ax1 = []

locVal    = GAN_v1['all_A_score'].values
locVal    = list(locVal)
locVal.sort(reverse = False) 
locVal    = np.asarray(locVal)

locVal  = np.asarray(locVal)

axs     =  plt.subplot(gs[0,0])
ax1     = axs

t       = np.arange(1,len(locVal)+1,1)

data1   = locVal

color   = 'tab:blue'
ax1.set_ylabel('Class Confidence', color=color, fontsize= 12)
ax1.plot(t, data1, color=color, linestyle='-', label='WGAN-GP Batch Size 8')
ax1.tick_params(axis='y', labelcolor=color,labelsize= 12)
base_mean = [np.mean(data1)]*len(t)
mean_line = ax1.plot(t,base_mean, linestyle=':', color='blue')

MeanValues = "Test Avg: "
axs.text(.1,.65,MeanValues,horizontalalignment='left',transform=axs.transAxes,fontsize= 12,color='black')

MeanValues = "WGAN-GP: Batch Size 8  = %.2f" % (base_mean[0])
axs.text(.1,.60,MeanValues,horizontalalignment='left',transform=axs.transAxes,fontsize= 12,color='blue')

####################################################################

locVal    = CVAE_v1['all_A_score'].values
locVal    = list(locVal)
locVal.sort(reverse = False) 
locVal    = np.asarray(locVal)

locVal  = np.asarray(locVal)

axs     =  plt.subplot(gs[0,0])
ax1     = axs

t       = np.arange(1,len(locVal)+1,1)

data1   = locVal

color   = 'tab:red'
ax1.set_ylabel('Class Confidence', color=color, fontsize= 12)
ax1.plot(t, data1, color=color, linestyle='-', label='WGAN-GP Batch Size 32')
ax1.tick_params(axis='y', labelcolor=color,labelsize= 12)
base_mean = [np.mean(data1)]*len(t)
mean_line = ax1.plot(t,base_mean, linestyle=':', color='red')
MeanValues = "WGAN-GP: Batch Size 32 = %.2f" % (base_mean[0])
axs.text(.1,.55,MeanValues,horizontalalignment='left',transform=axs.transAxes,fontsize= 12,color='red')

####################################################################

locVal    = CVAE_v3['all_A_score'].values
locVal    = list(locVal)
locVal.sort(reverse = False) 
locVal    = np.asarray(locVal)

locVal  = np.asarray(locVal)

axs     =  plt.subplot(gs[0,0])
ax1     = axs

t       = np.arange(1,len(locVal)+1,1)

data1   = locVal

color   = 'tab:green'
ax1.set_ylabel('Class Confidence', color='black', fontsize= 12)
ax1.plot(t, data1, color=color, linestyle='-', label='CVAE (From Scratch)')
ax1.tick_params(axis='y', labelcolor='black',labelsize= 12)
base_mean = [np.mean(data1)]*len(t)
mean_line = ax1.plot(t,base_mean, linestyle=':', color='green')

ax1.legend(loc='upper left', fontsize= 12)
ax1.set_xlabel('Number of Images', color='black', fontsize= 12)

MeanValues = "CVAE (From Scratch) = %.2f" % (base_mean[0])
axs.text(.1,.50,MeanValues,horizontalalignment='left',transform=axs.transAxes,fontsize= 12,color='green')

# ####################################################################

locVal    = Val_Set['all_A_score'].values
locVal    = list(locVal)
locVal.sort(reverse = False) 
locVal    = np.asarray(locVal)

locVal  = np.asarray(locVal)

axs     =  plt.subplot(gs[0,0])
ax1     = axs

t       = np.arange(1,len(locVal)+1,1)

data1   = locVal

color   = 'magenta'
ax1.set_ylabel('Class Confidence', color='black', fontsize= 12)
ax1.plot(t, data1, color=color, linestyle='-', label='CVAE (From Existing)')
ax1.tick_params(axis='y', labelcolor='black',labelsize= 12)
base_mean = [np.mean(data1)]*len(t)
mean_line = ax1.plot(t,base_mean, linestyle=':', color='magenta')

ax1.legend(loc='upper left', fontsize= 12)
ax1.set_xlabel('Number of Images', color='black', fontsize= 12)

MeanValues = "CVAE (From Existing) = %.2f" % (base_mean[0])
axs.text(.1,.45,MeanValues,horizontalalignment='left',transform=axs.transAxes,fontsize= 12,color='magenta')

```
